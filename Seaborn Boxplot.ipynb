Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used for detect the outlier in data set. It captures the summary of the data efficiently with a simple box and whiskers and allows us to compare easily across groups. Boxplot summarizes a sample data using 25th, 50th and 75th percentiles. These percentiles are also known as the lower quartile, median and upper quartile.

A box plot consist of 5 things.

* Minimum
* First Quartile or 25%
* Median (Second Quartile) or 50%
* Third Quartile or 75%
* Maximum

```python
# it is a distribution plot for the categorical varaible
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Box plot gives distribution of data interms of summary statistics
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
# filter cars with even number of cylinders
cars = cars[cars['cylinders'].isin([2,4,6,8])]
```

```python
sns.boxplot(cars['mpg'])
```

```python
cars['mpg'].describe()
```

```python
sns.boxplot(x = cars['origin'], y = cars.mpg)
# origin is categorical variable
```

```python
sns.boxplot(x = 'origin', y = 'mpg', data = cars)
```

```python
### Hue
# by cylinders
sns.boxplot(x = 'origin', y = 'mpg', data = cars, hue = 'cylinders')
```

```python
## by model year
cars.model_year.describe()
```

```python
# creating a new column
cars['newer_model'] = cars.model_year > 76
# this will give the true or false in new column
```

```python
sns.boxplot(x = 'origin', y = 'mpg', data = cars, hue = 'newer_model')
```

```python
## styling
```

```python
sns.boxplot(y = 'origin', x = 'mpg', data = cars,)
```

```python
# ordering the bars of box plot
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           order = ['japan', 'europe', 'usa'] )
```

```python
# change the order of hue
sns.boxplot(data = cars,x = 'origin', y = 'mpg', order = ['japan', 'europe', 'usa'],hue = 'newer_model', hue_order = [True, False])
```

```python
### color
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           color = 'g' )
```

```python
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           color = 'g', hue = 'newer_model' )
```

```python
### width of box plot
sns.boxplot(data = cars,y = 'origin', x = 'mpg',
           width = 0.5 )
```

```python
### linewidth
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           linewidth = 2.3 )
```

```python
### whis for whisker
# We use 1.5 times IQR for whiskers but we can change it
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           whis = 1 )
```

```python
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           whis = 2 )
```

```python
#The outliers will change when we use whis parameter
```

```python
## fliersize
# these are the little diamonds used to represent the outlier
# we can change its size
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           fliersize = 2 )
```

```python
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           fliersize = 8 )
```

```python
## showcaps
# this will remove the caps or lines
sns.boxplot(data = cars,x = 'origin', y = 'mpg',
           showcaps = False )
```
