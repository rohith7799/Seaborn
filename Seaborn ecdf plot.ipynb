* its a distribution plot like histplot and displot
* Stands for empirical cumulative distribution function
* ECDF stands for Empirical Commutative Distribution. 
* It is more likely to use instead of the histogram for visualizing the data because the ECDF plot visualizes each and every data point of the dataset directly, which makes it easy for the user to interact with the plot.
* This plot contains more information because it has no bin size setting, which means it doesnâ€™t have any smoothing parameters.
* Since its curves are lines and monotonically increasing, so it is well suited for comparing multiple distributions at the same time.
* In an ECDF plot, the x-axis corresponds to the range of values for the variable whereas the y-axis corresponds to the proportion of data points that are less than or equal to the corresponding value of the x-axis.
* We can make the ECDF plot directly by using ecdfplot() function, or we can also make the plot by using displot() function with the new Seaborn version.

disadvantage
* cannot view the central tendencies

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
#### Basic
```

```python
sns.set_style('darkgrid')
```

```python
sns.ecdfplot(x = 'tip', data = tips)
```

```python
tips.tip.min()
```

```python
tips.tip.max()
```

```python
# where ever there are straigt vertical lines it means that value
# is repeated

tips.tip.value_counts()
```

```python
# We can see what percent of distribution is above or below some
# value
sns.ecdfplot(x = 'tip', data = tips)
plt.axvline(4, c= 'black') 
# the y axis represent the proportion of the data.
# the line is drawn at 4 (x axis is tips in dollars) and it points
# at 0.8. this means 80 % of the tips is below 4$
```

```python
#### Additional options
```

```python
# hue for categorical variable
sns.ecdfplot(x = 'tip', data = tips, hue = 'time')
```

```python
sns.ecdfplot(x = 'tip', data = tips, hue = 'day')
```

```python
### Count
# so far we have seen proportion on y axis, now let us see count

sns.ecdfplot(x = 'tip', data = tips, stat = 'count', hue = 'day')
```

```python
sns.ecdfplot(x = 'tip', data = tips, stat = 'count')
```

```python
# Rank in the plot
# 
sns.ecdfplot(x = 'tip', data = tips, stat = 'count', complementary = True)
```

```python
#the no.1 best tip is 10$
```

```python
### weights
# use weights to weight the proportion. here 50% of tip money made came
# from tips less than 3.5$

sns.ecdfplot(x = 'tip', data = tips, weights = 'tip')
```

```python
# when i am at 2 on x axis, y is 0.2. this means by the time i reach
# 2 i would make 20% or 0.2 of total money i am going to make
```

```python
### Styling
```

```python
### Palette
sns.ecdfplot(x = 'tip', data = tips, hue = 'day', palette = 'summer')
```

```python
#### line width, lw
p = sns.ecdfplot(x = 'tip', data = tips, hue = 'day', 
             palette = 'summer', lw = 3)
# changing the legend and legend title and legend location
p.legend(['thursday', 'friday', 'saturday', 'sunday'], title = 'day of week', loc = 4)
```
