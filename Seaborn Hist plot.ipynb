```python
# Histogram using histplot in seaborn
# previously used to build histogram from distplot but in updated
# seaborn we have histplot

# we have histplot, displot, ecdfplot for ploting the distributions
```

```python
import seaborn as sns
```

```python
sns.__version__
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
penguins.dropna(inplace = True )
```

```python
penguins.shape
```

```python
## Intro Visuals
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(penguins.bill_length_mm)
```

```python
# we can pass the entire dataframe and just pass the 
# column name in the x variable
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins)
```

```python
# Horizontal histogram

sns.histplot(y = 'bill_length_mm', data = penguins)
```

```python
# KDE plot
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, 
            kde = True)
```

```python
# By default histogram decides the no.of bins. 
# but we can specify the no.of bins

sns.histplot(x = 'bill_length_mm', data = penguins, bins = 20)

```

```python
# We can choose the bins to be spaced irregularly by passing 
# the list in the bins argument

sns.histplot(x = 'bill_length_mm', data = penguins,
            bins = [30,40,45,50,60])

# the list of number specifies the start position of the bins
```

```python
# Binwidth and Binrange
# Binwidth helps to choose the widht of bins. 
# if binwidth = 10 then the width of each bin will be 10

sns.histplot(x = 'bill_length_mm', data = penguins, 
            binwidth = 5)
```

```python
# Binrange helps us to specify the range for the bins

sns.histplot(x = 'bill_length_mm', data = penguins, 
            binrange = [30, 60])

# start of the bins is from 30 and end is at 60
```

```python
# Statistics argument
# we can change the statistics of the graph. by defualt we
# have the count stat (what is plotted on y axis) but we can change it
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            stat = 'density')
#  density =  area of histogram sums to one
# if calculate product of height and width of each bar and 
# sum it up, we will get sum to be 1 which is similar to area under pdf
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            stat = 'probability')
# stat = probability will tell us the probability
```

```python
# probability cummulative graph

sns.histplot(x = 'bill_length_mm', data = penguins,
            stat = 'probability',
            fill = False,
            element = 'step',
            cumulative = True)

# This will give the cumulative distribution
# element = step will give us the step instead of bars
```

```python
# Dealing categorical variables
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species')

# we pass the categorical variable into hue parameter
# we can see the histogram for each species
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species', element = 'step')
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species', element = 'poly')
```

```python
# multiple parameter
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species', multiple = 'layer')
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species', multiple = 'stack')
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins,
            hue = 'species', multiple = 'fill')
# multiple = fill will plot y axis between 0 to 1 and will tell
# the proportion of each category
```

```python
# Bivariate histogram
# we plot 2 differnet variables
```

```python
sns.set_style('white')
```

```python
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm',
            data = penguins,
            )
```

```python
# color bar = it will put the color bar with varing density 
# on the right side of the graph
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm',
            data = penguins,
             cbar = True
            )
```

```python
# hue parameter for categorical varibale
sns.histplot(x = 'bill_length_mm', y = 'bill_depth_mm',
            data = penguins,
             hue = 'species'
            )
```

```python
sns.histplot(x = 'bill_length_mm', y = 'species',
            data = penguins,
             hue = 'species'
            )
```

```python
# styling
# shrink parameter
sns.histplot(x = 'species', hue = 'sex',
            multiple = 'dodge',
            data = penguins,
            shrink = 0.8)
```

```python
# color, palette

sns.histplot(x = 'bill_length_mm', data = penguins,
           color = 'indigo', 
            palette = 'bone')
```
