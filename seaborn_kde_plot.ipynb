```python
# to understand the distribution of the data i.e, smoother version
# histogram also gives the distribution which is discrete in nature and not smooth
# KDE plot gives the probability density function from the data set
```

```python
# Seaborn univariate kde plot
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.kdeplot(cars['horsepower'])
```

```python
# shade parameter = shades the graph
```

```python
sns.kdeplot(cars['horsepower'], shade = True)
```

```python
# bandwidth parameter
```

```python
sns.kdeplot(cars['horsepower'], bw_adjust = 10)

# if bw is low then there are more peaks, if increased the curve
# smooth out and looses the peaks
```

```python
# cummulative: if true will gives the cummulative distribution.
# the resultant graph is the cumulative distribution plot instead
# of probability distribution plot
```

```python
sns.kdeplot(cars['horsepower'], cumulative = True)
```

```python
# Bivariate KDE plot - gives the joint pdf of the variables
```

```python
sns.kdeplot(cars.horsepower, cars.mpg)
```

```python
# n_levels changes the no.of circles in the plot
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels = 20)
```

```python
shading the grapth
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, shade = True)
```

```python
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg,
           shade = True,
           shade_lowest = False)
```

```python
# cbar = color bar to see the densities
```

```python
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg,
           shade = True,
           shade_lowest = False,
           cbar = True)
```

```python
sns.set_style('whitegrid')

sns.kdeplot(cars.horsepower, cars.mpg,
           shade = True,
           hue = cars.cylinders)
```
